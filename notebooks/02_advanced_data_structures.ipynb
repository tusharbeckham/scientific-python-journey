## 1. Lists
* **Definition:** Lists are ordered, mutable (changeable) collections. In machine learning, lists are often your first step in gathering sequential data before moving it into more advanced structures like Pandas DataFrames.

```python
# Creating and Indexing
solar_data = [250, 275, 300, 290]
print(solar_data[0])    # 250 (First item)
print(solar_data[-1])   # 290 (Last item)
print(solar_data[1:3])  # [275, 300] (Slicing: up to but not including index 3)
```

* **Key Methods:** `.append()`, `.insert()`, `.remove()`, `.pop()`

```python
solar_data = [250, 275, 300, 290]
solar_data.append(310)       # Adds to the end
solar_data.insert(1, 260)    # Inserts 260 at index 1
solar_data.remove(300)       # Removes the first occurrence of 300
popped_val = solar_data.pop() # Removes and returns the last item
print(popped_val)          # 310
print(solar_data)          # [250, 260, 275, 290]
```

List Comprehensions (Crucial for AI/ML):
A highly efficient way to create lists in a single line.

```python
# Standard way to square numbers
squares = []
for x in range(5):
    squares.append(x**3)

# Comprehension (Faster and cleaner)
squares_comp = [x**3 for x in range(5)]
print(squares)       # [0, 1, 8, 27, 64]
print(squares_comp)  # [0, 1, 8, 27, 64]
```

## 2. Tuples
* **Definition:** Tuples look like lists but use parentheses (). The critical difference: Tuples are immutable. Once created, you cannot add, remove, or change elements.

Why use them? They are slightly faster and more memory-efficient than lists. Use them for data that should not change (like geographic coordinates or model hyperparameters).

```python
# Creating and Unpacking
dimensions = (1920, 1080)

# Tuple Unpacking (very common when functions return multiple values)
width, height = dimensions
print(f"Width: {width}, Height: {height}")
```

## 3. Dictionaries
* **Definition:** Dictionaries use {} and store data in key:value pairs. If you are building a web app, this is exactly how JSON data flows between your backend and frontend.

* **Key Methods:** `.keys()`, `.values()`, `.get()`

```python
# Creating a dictionary
panel = {
    "brand": "SunPower",
    "capacity": 400,
    "efficiency": 0.22
}

# Accessing and Modifying
print(panel["brand"])          # "SunPower"
panel["location"] = "Roof A"   # Adds a new key-value pair
panel["capacity"] = 410        # Updates an existing key

# Dictionary Methods
print(panel.keys())            # dict_keys(['brand', 'capacity', 'efficiency', 'location'])
print(panel.values())          # dict_values(['SunPower', 410, 0.22, 'Roof A'])

# Safe access with .get() - prevents errors if the key doesn't exist
print(panel.get("warranty", "Not Found")) # Returns "Not Found" instead of crashing
```

## 4. Sets
* **Definition:** Sets use {} like dictionaries, but they only contain unique values and are unordered. They are mathematically powerful and incredibly fast for checking if an item exists.

```python
# Removing duplicates from a list
raw_data = ["Delhi", "Mumbai", "Delhi", "Pune", "Mumbai"]
unique_cities = set(raw_data)
print(unique_cities) # {'Pune', 'Mumbai', 'Delhi'}

# Set Math
group_a = {"React", "Python", "SQL"}
group_b = {"Python", "PHP", "SQL"}

print(group_a.intersection(group_b)) # What's in both? {'Python', 'SQL'}
print(group_a.difference(group_b))   # What's only in A? {'React'}
```

## 5. The Collections Module

* **`Counter`:** A supercharged dictionary designed specifically for counting the frequency of elements in a list or iterable.
* **`defaultdict`:** A dictionary that automatically assigns a default value (like `0` or an empty list `[]`) if you try to access a key that doesn't exist yet, preventing `KeyError` crashes.
* **`namedtuple`:** Creates a tuple where you can access the data using descriptive names (like `student.math_score`) instead of confusing indexes (like `student[2]`).

```python
from collections import Counter

grades = ['A', 'B', 'A', 'C', 'B', 'A', 'F', 'B']
grade_counts = Counter(grades)

print(grade_counts)        # Output: Counter({'A': 3, 'B': 3, 'C': 1, 'F': 1})
print(grade_counts['A'])   # Output: 3
```

```python
from collections import defaultdict

# We tell it: "If a key doesn't exist, make its value an empty list []"
grouped_students = defaultdict(list)

# We can immediately start appending without checking if the key exists!
grouped_students['Excellent'].append("Aayush")
grouped_students['Needs Improvement'].append("Rohan")
grouped_students['Excellent'].append("Piyush")

# Output: {'Excellent': ['Aayush', 'Piyush'], 'Needs Improvement': ['Rohan']}
print(dict(grouped_students))
```

```python
from collections import namedtuple

# 1. Define the "blueprint"
StudentRecord = namedtuple('Student', ['name', 'attendance', 'final_score'])

# 2. Create a record
student_1 = StudentRecord("Tushar", 92, 99)

# 3. Access data cleanly by name instead of index!
print(student_1.name)         # Output: Tushar
print(student_1.final_score)  # Output: 99
```

```python
# --- EXERCISE 4: COLLECTIONS MODULE ---
from collections import Counter, defaultdict, namedtuple

# 1. Counter: Find out how many times each subject appears in this list
favorite_subjects = ["Math", "Physics", "Math", "Computer Science", "Physics", "Math"]
# Write your Counter code here:
counter_subjects = Counter(favorite_subjects)
print("Subject Counts:", counter_subjects)

# 2. defaultdict: Group these scores into a dictionary
# Keys should be the subject, values should be a list of scores
raw_scores = [("Math", 88), ("Physics", 75), ("Math", 92), ("Computer Science", 95)]
# Write your defaultdict code here:
grouped_scores = defaultdict(list)
for subject, score in raw_scores:
    grouped_scores[subject].append(score)
print("Grouped Scores:", dict(grouped_scores))


# 3. namedtuple: Create a 'ServerConfig' namedtuple with fields: 'host', 'port', 'status'
# Then create one instance of it and print the port number.
# Write your namedtuple code here:
ServerConfig = namedtuple('ServerConfig', ['host', 'port', 'status'])
server_1 = ServerConfig(host="localhost", port=8080, status="running")
print("Server Port:", server_1.port)
```
