## Lists, Dictionaries & Loops

```python
# Exercise 4
temperatures = [28, 31, 29, 35, 33, 30, 27]

print("Max Temp:", max(temperatures))
print("Min Temp:", min(temperatures))
print("Average Temp:", round(sum(temperatures)/len(temperatures), 2))

# List Comprehension
hot_days = [temp for temp in temperatures if temp > 30]
print("Hot days (>30°C):", hot_days)
```

```python
# Exercise 5
solar_panel = {
    "brand": "Longi",
    "capacity": 550,
    "efficiency": 21.5,
    "location": "Delhi",
    "installed": "2025"
}

print("Brand:", solar_panel["brand"])
print("Capacity:", solar_panel["capacity"], "watts")
print("Efficiency:", solar_panel["efficiency"], "%")

# Adding new key
solar_panel["daily_output"] = 4.8
print("Daily Output:", solar_panel["daily_output"], "kWh")
```

```python
# Exercise 6
def calculate_power_efficiency(power_output, area):
    """Calculate efficiency of solar panel"""
    irradiance = 1000  # standard value
    efficiency = (power_output / (irradiance * area)) * 100
    return round(efficiency, 2)

print("Efficiency of Panel:", calculate_power_efficiency(450, 2.5), "%")
```
